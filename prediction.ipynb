```python
import pandas as pd
data = pd.read_csv('./DATASET.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
def prediction (row):
    if row['BMI'] >= 25:
        return 'Y'
    elif row['Chol'] > 5:
        return 'Y'
    else:
        return 'N'
data['pred']=data.apply(lambda row: prediction(row), axis = 1)
def TP (row):
    if row['CLASS'] == 'Y' and row['pred'] == 'Y':
        return 1
    else:
        return 0
data['TP']= data.apply(lambda row: TP(row), axis = 1)

def FN (row):
    if row['CLASS'] == 'Y' and row['pred'] == 'N':
        return 1
    else:
        return 0
data['FN']= data.apply(lambda row: FN(row), axis = 1)

def TN (row):
    if row['CLASS'] == 'N' and row['pred'] == 'N':
        return 1
    else:
        return 0
data['TN']= data.apply(lambda row: TN(row), axis = 1)

def FP (row):
    if (row['CLASS'] == 'N' and row['pred'] == 'Y'):
        return 1
    else:
        return 0
data['FP']= data.apply(lambda row: FP(row), axis = 1)
```

```python
count_tp = data['TP'].value_counts()[1]
count_fn = data['FN'].value_counts()[1]
count_tn = data['TN'].value_counts()[1]
count_fp = data['FP'].value_counts()[1]
total = count_tp+count_fn+count_tn+count_fp
print(count_tp,count_fn,count_tn,count_fp)
print(total)
accuracy = (count_tp + count_tn) / total * 100
print(accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
number = LabelEncoder()
data['ID'] = number.fit_transform(data['ID'])
data['No_Pation'] = number.fit_transform(data['No_Pation'])
data['Gender'] = number.fit_transform(data['Gender'])
data['AGE'] = number.fit_transform(data['AGE'])
data['Urea'] = number.fit_transform(data['Urea'])
data['Cr'] = number.fit_transform(data['Cr'])
data['HbA1c'] = number.fit_transform(data['HbA1c'])
data['TG'] = number.fit_transform(data['TG'])
data['HDL'] = number.fit_transform(data['HDL'])
data['LDL'] = number.fit_transform(data['LDL'])
data['VLDL'] = number.fit_transform(data['VLDL'])
data['BMI'] = number.fit_transform(data['BMI'])
data['CLASS'] = number.fit_transform(data['CLASS'])
```

```python
features = ["ID","No_Pation","Gender", "AGE", "Urea", "Cr", "HbA1c","Chol","TG","HDL","LDL","VLDL","BMI"]
target = "CLASS"
features_train, features_test, target_train, target_test = 
train_test_split(data[features],data[target],test_size = 0.4,random_state = 54)
```

```python
model = GaussianNB()
model.fit(features_train, target_train)
```

```python
pred = model.predict(features_test)
accuracy = accuracy_score(target_test, pred)
accuracy * 100
```
